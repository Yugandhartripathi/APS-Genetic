```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('GAdataFromgForm.csv')
```

```python
df.head()
```

```python
df.drop('Timestamp',inplace=True,axis=1)
```

```python
AOI=df['Areas of Interests']
```

```python
df['Interest_WebDev']=np.zeros(shape=(df.shape[0],),dtype=int)
df['Interest_Android/IOS']=np.zeros(shape=(df.shape[0],),dtype=int)
df['Interest_GameDev']=np.zeros(shape=(df.shape[0],),dtype=int)
df['Interest_ML']=np.zeros(shape=(df.shape[0],),dtype=int)
df['Interest_Security']=np.zeros(shape=(df.shape[0],),dtype=int)
df['Interest_IoT']=np.zeros(shape=(df.shape[0],),dtype=int)
```

```python
AOI[0]
```

```python
for i in range(AOI.shape[0]):
    interests=AOI[i]
    interest=interests.split(' ')
    for x in interest:
        if x=='Web':
            df['Interest_WebDev'][i]=1
        elif x=='Android':
            df['Interest_Android/IOS'][i]=1
        elif x=='Game':
            df['Interest_GameDev'][i]=1
        elif x=='Machine':
            df['Interest_ML'][i]=1
        elif x=='Cyber':
            df['Interest_Security'][i]=1
        elif x=='Internet':
            df['Interest_IoT'][i]=1
```

```python
df.drop('Areas of Interests',axis=1,inplace=True)
```

```python
df.head()
```

```python
textEditors=pd.get_dummies(df['Preferred Text Editor'],prefix='PrefTE',prefix_sep='_')
```

```python
OS=pd.get_dummies(df['Preferred OS '],prefix='PrefOS',prefix_sep='_')
```

```python
df=pd.concat([df,textEditors,OS],axis=1)
```

```python
df.drop(['Preferred OS ','Preferred Text Editor'],axis=1,inplace=True)
```

```python
df.columns
```

```python
for i in range(df.shape[0]):
    temp=[]
    temp=(df['Name'][i].split(' '))
    df['Name'][i]=temp.join();
```

```python
df.to_csv('processedData',index=False)
```
